```python
#plotting function
'''
#alternative option to find the R^2 value using user defined function
def rsq(Y,yresid):
    SSresid = np.sum(yresid**2)
    SStotal = len(y) * np.var(y)
    rsq = 1 - SSresid/SStotal
    print('R^2 = ',rsq)
    return rsq
'''
titles=["x(m) vs Electric Potential(Voc)","x(m) vs Electric charge(pC)","x(m) vs Capacitance(pF)"]
ylabels=["Electric Potential(Voc)","Electric charge(pC)","Capacitance(pF)"]
#plt.plot(X, np.sin(X), label = "Curve-1")
#plt.title("Line Graph - Geeksforgeeks")

def ploting(x,y,titles,ylabels):
    
    #polyfit
    p1 = np.polyfit(x,y,1)
    p2 = np.polyfit(x,y,2)
    p3 = np.polyfit(x,y,3)
    p4 = np.polyfit(x,y,4)
    p5 = np.polyfit(x,y,5)
    p6 = np.polyfit(x,y,6)
    
    #finding yresiduals
    yfit1=np.polyval(p1,x)
    yfit2=np.polyval(p2,x)
    yfit3=np.polyval(p3,x)
    yfit4=np.polyval(p4,x)
    yfit5=np.polyval(p5,x)
    yfit6=np.polyval(p6,x)
    '''
    #finding yresiduals (alternative option/inside the polyval method)
    yfit1 = p1[0] * x    + p1[1]
    yfit2 = p2[0] * x**2 + p2[1]*x    + p2[2]
    yfit3 = p3[0] * x**3 + p3[1]*x**2 + p3[2]*x   + p3[3]
    yfit4 = p4[0] * x**4 + p4[1]*x**3 + p4[2]*x**2 + p4[3]*x   + p4[4]
    yfit5 = p5[0] * x**5 + p5[1]*x**4 + p5[2]*x**3 + p5[3]*x**2 + p5[4]*x + p5[5]
    yfit6 = p6[0] * x**6 + p6[1]*x**5 + p6[2]*x**4 + p6[3]*x**3 + p6[4]*x**2 + p6[5]*x + p6[6]
    '''
    yfits=[yfit1,yfit2,yfit3,yfit4,yfit5,yfit6]
    #print(yfits)
    
    #plots using matplotlib
    import matplotlib.pyplot as plt
    plt.rcParams['figure.figsize'] = [15, 8]
    #plt.rcParams['figure.dpi'] =200
    
    plt.plot(x,y,'o-',label='Actual data', linewidth=2, markersize=8)
    #x = np.linspace(-2,6,100)
    plt.plot(x,yfit1,'y-',label='yfit1', linewidth=2, markersize=8)
    plt.plot(x,yfit2,'b--',label='yfit2', linewidth=2, markersize=8)
    plt.plot(x,yfit3,'k:',label='yfit3', linewidth=2, markersize=8)
    plt.plot(x,yfit4,'c-',label='yfit4', linewidth=2, markersize=8)
    plt.plot(x,yfit5,'m:',label='yfit5', linewidth=2, markersize=8)
    plt.plot(x,yfit6,'r--',label='yfit6', linewidth=2, markersize=8)
    
    plt.title(titles)
    plt.xlabel("Seperation distance x(m)")
    plt.ylabel(ylabels)
    
    #finding R^2 values and printing the equations
    from sklearn.metrics import r2_score
    print('='*100)
    print(p1)
    P1=np.poly1d(p1)
    print(P1)
    print('R^2_1= ',r2_score(y,yfits[0]))
    print('='*100)
    
    print(p2)
    P2=np.poly1d(p2)
    print(P2)
    print('R^2_2= ',r2_score(y,yfits[1]))
    print('='*100)
    
    print(p3)
    P3=np.poly1d(p3)
    print(P3)
    print('R^2_3= ',r2_score(y,yfits[2]))
    print('='*100)
    
    print(p4)
    P4=np.poly1d(p4)
    print(P4)
    print('R^2_4= ',r2_score(y,yfits[3]))
    print('='*100)
    
    print(p5)
    P5=np.poly1d(p5)
    print(P5)
    print('R^2_5= ',r2_score(y,yfits[4]))
    print('='*100)
    
    print(p6)
    P6=np.poly1d(p6)
    print(P6)
    print('R^2_6= ',r2_score(y,yfits[5]))
    print('='*100)
    plt.legend()
    plt.show()
    

```
